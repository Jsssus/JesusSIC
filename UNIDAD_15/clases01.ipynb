Algoritmo Naive Bayes

```python
#Algorito de Naive Bayes

from sklearn.datasets import load_iris
import seaborn as sns
import pandas as pd
```

```python
datos = load_iris()

print(datos.DESCR)
```

```python
#dividir los datos
x = datos['data'] #datos.data es igual
colum = list(datos['feature_names'])
```

```python
y = datos['target']
especies = list(datos['target_names'])
```

```python
import matplotlib.pyplot as plt

var_y = pd.Series(y)
tabla = var_y.value_counts()
tabla = tabla.sort_index()

sns.barplot(tabla.values)
plt.show()
```

```python
var_x = pd.DataFrame(x, columns=['Sepal_L', 'Sepal_W', 'Petal_L', 'Petal_W'])
my_cols_dict = {0 : 'red', 1 : 'green', 2 : 'blue'}
my_cols = pd.Series(y).apply(lambda x: my_cols_dict[x])
pd.plotting.scatter_matrix(var_x,  c=my_cols, marker='^', alpha=0.5)
plt.show
```

```python
from sklearn.model_selection import train_test_split #importar el selector para dividir los datos

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
from sklearn.naive_bayes import GaussianNB

GNB = GaussianNB()
GNB.fit(x_train, y_train)
y_pred_test = GNB.predict(x_test)
```

```python
from sklearn import metrics

confusion_m = metrics.confusion_matrix(y_test, y_pred_test)
print(confusion_m)
```

```python
import numpy as np

prec = metrics.accuracy_score(y_test, y_pred_test)
print('accuracy: ', str(np.round(prec,3)))
prec
```
