````python
# Obtener tiempo actual

import time

t_act = time.time() # 1 de enero de 1970, 00:00:00 UTC ## The epoch is the point where the time starts
# La época actual del 1 de enero de 1970 a las 00:00:00 UTC fue seleccionada arbitrariamente por los 
#ingenieros de Unix porque se consideró una fecha conveniente para trabajar 
print("Tiempo actual en segundos", t_act)
````

````python
# Conversión de tiempo

import time

seg = 1709680821.2168674
t_tupla = time.gmtime(seg)
print("Tiempo en formato de tupla", t_tupla)
````

````python
# Manipulación del objeto datetime
import datetime

f_act = datetime.datetime.now()

print("Año: ", f_act.year)
print("Mes: ", f_act.month)
print("Día: ", f_act.day)
print("Hora: ", f_act.hour)
print("Minuto: ", f_act.minute)
print("Segundo: ", f_act.second)

# Sumar una fecha futura
print("_____________")
duracion = datetime.timedelta(days=7)
f_futura = f_act + duracion
print("Fecha dentro de una semana: ", f_futura)

# Reemplazar año y mes
print("_____________")
f_modf = f_act.replace(year=2025, month=8)
print("Fecha modificada: ", f_modf)

# Formato a un obj datetime
print("_____________")
print("Objeto Date time", f_act)
f_formato = f_act.strftime("%d-%m-%Y %H:%M:%S")
print("Fecha con formato: ", f_formato)

# Pasar una str a un obj datetime
f_str = "2025-01-16"
f_obj = datetime.datetime.strptime(f_str, "%Y-%m-%d")
print("Fecha objeto: ", f_obj)
````

````python
import time

## Uso de ctime
t_seg = time.time()
f_h_legible = time.ctime(t_seg)
print("Fecha y hora legibles: ", f_h_legible)

## Uso de gmtime
print("_____________")
t_utc = time.gmtime(t_seg)
print("Tiempo en UTC: ", t_utc)
````

df.dropna(): Elimina filas o columnas con valores nulos.

````python
x = [1,2,3,4,5,6]
y = [4,3,5,6,7,4]

fig = plt.figure()

subplot1 = fig.add_subplot(2,1,1)
subplot1.plot(x, y)

subplot2 = fig.add_subplot(2,1,2)
subplot2.plot(x, y, '-.d')
subplot2.text(4,5, '2do subplot')

fig.suptitle("Add subplots to a figure")
plt.show()
````

````python
import matplotlib.pyplot as plt

x = np.arange(0.0, 3.0, 0.01)

def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

ax1 = plt.subplot(212)
ax1.plot(x, f(x))

ax2 = plt.subplot(221)
ax2.margins(1,1)
ax2.plot(x,f(x))
ax2.set_title('Zoomed out')

ax3 = plt.subplot(222)
ax3.margins(x=0, y=-0.25) # Values in (-0.5, 0.0) zooms in to center
ax3.plot(x, f(x))
ax3.set_title('Zoomed in')

plt.show()
````
